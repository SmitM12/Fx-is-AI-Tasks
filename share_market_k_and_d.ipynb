```python
# Importing packages
import pandas as pd
```

```python
# load the dataset
data = pd.read_csv('TCS.csv')
data
```

```python
# Dropping unnecessary features
data = data.drop(['Symbol', 'Series', 'Prev Close', 'Open', 'Last', 'VWAP', 'Volume', 'Turnover', 'Trades', 'Deliverable Volume', '%Deliverble'], axis=1)
data
```

```python
data.isnull().value_counts()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data
```

```python
# Calculating %K
lookback_period = 14
data['Lowest Low'] = data['Low'].rolling(window=lookback_period).min()
data['Highest High'] = data['High'].rolling(window=lookback_period).max()
data['%K'] = (data['Close'] - data['Lowest Low']) / (data['Highest High'] - data['Lowest Low']) * 100
data
```

```python
# Calculating %D
data['%D'] = data['%K'].rolling(window=3).mean()
data
```

```python
data.head()
```

```python
# Display the result
print(data[['Close', '%K', '%D']].tail())
```
